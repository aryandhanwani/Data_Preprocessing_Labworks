Que 3:- Tensors
- Create a scalar, vector, matrix, and 3D tensor in Pandas and PyTorch.
- Print their shapes and dimensions.

```python
import pandas as pd
import torch 

#Scalar 
scalar = torch.tensor(5)

#Vector (1D)
vector = torch.tensor([1, 2, 3])

#Matrix (2D)
matrix = torch.tensor([[1, 2], [3, 4]])

# 3D Tensor 
tensor_3d = torch.tensor([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])

# Print shapes
print("Scalar shape:", scalar.shape)
print("Vector shape:", vector.shape)
print("Matrix shape:", matrix.shape)
print("3D Tensor shape:", tensor_3d.shape)

# Dimensions
print("\nDimensions:")
print("Scalar dim:", scalar.dim())
print("Vector dim:", vector.dim())
print("Matrix dim:", matrix.dim())
print("3D Tensor dim:", tensor_3d.dim())
```

Que 4:- Tensors
- Generate a random tensor (shape = (3,3,3)).
Perform:
- Reshape to (9,3).
Slice out the first two "layers".
- Matrix multiplication between two 2D tensors.
- Demonstrate broadcasting with two tensors of shapes (3,1) and (3,3).

```python
#Create tensor 
x= torch.tensor([[1, 2, 3], [4, 5, 6]])

#Reshaping
reshaped = x.reshape(3,2)
print("Reshaped:\n", reshaped)

#Slicing 
slice_part = x[:, 1]
print("\nSliced:\n", slice_part)

#Broadcasting
y = torch.tensor([10, 20, 30])
result = x + y
print("\nBroadcasting Result:\n", result)
```

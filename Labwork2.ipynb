Que 1:- Working with CSV File
Download or create a CSV file with student details (Name, Age, Marks).
Load it into a Pandas DataFrame.
- Display first 5 rows, column names, and shape.
- Filter students with marks > 70.
- Save the filtered DataFrame into a new CSV file.

```python
import pandas as pd

# Create DataFrame
data = {
    "Name": ["Aryan", "Rohit", "Neha", "Priya", "Aman"],
    "Age": [20, 22, 19, 21, 23],
    "Marks": [75, 60, 85, 90, 65]
}
df = pd.DataFrame(data)

# Save CSV
df.to_csv("students.csv", index=False)

# Load CSV
df = pd.read_csv("students.csv")

# Display
print(df.head())
print(df.columns)
print(df.shape)

# Filter
filtered = df[df["Marks"] > 70]

# Save filtered
filtered.to_csv("filtered_students.csv", index=False)
```

Que 4:- Fetching Data from an API
- Use the REST Countries API:
https://restcountries.com/v3.1/all?fields=name,flags
- Fetch data with requests.getO.
- Convert JSON response into a Pandas DataFrame.
- Extract country name, capital, and population into a new DataFrame.

```python
import requests
import pandas as pd

url = "https://restcountries.com/v3.1/all?fields=name,capital,population"

response = requests.get(url)
data = response.json()

# Convert to DataFrame
df = pd.DataFrame(data)

# Extract required fields
new_df = pd.DataFrame({
    "Country": df["name"].apply(lambda x: x["common"]),
    "Capital": df["capital"],
    "Population": df["population"]
})

print(new_df.head())
```

```python
from sklearn.preprocessing import MinMaxScaler

# Create sample data with missing values
data = {
    "Age": [20, 25, None, 30, 35],
    "Marks": [80, None, 70, 60, 90]
}
df = pd.DataFrame(data)

# Display info
print(df.info())
print(df.describe())

# Handle missing values
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Marks"].fillna(df["Marks"].median(), inplace=True)

# Drop duplicates
df.drop_duplicates(inplace=True)

# Normalize
scaler = MinMaxScaler()
df["Marks"] = scaler.fit_transform(df[["Marks"]])

print(df)
```

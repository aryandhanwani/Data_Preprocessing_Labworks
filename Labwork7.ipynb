Q.1 : Outliers in Machine Learning

Theory Recap:
- Outliers = data points that deviate significantly from the rest of the dataset.
- Can distort:
- Mean, variance, correlation
- Model training (especially regression, distance-based models,
clustering, etc.)
- Handling: detect, analyze, remove or cap.

Tasks:
- Generate a dataset of 100 normal-distributed values for Age.
- Manually add 5 extreme values (200, 150, -10, 170, 300).
- Plot histogram and boxplot to visualize outliers.
- Comment on how outliers affect mean vs median.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
age = np.random.normal(loc=30, scale=5, size=100).astype(int)
df = pd.DataFrame({'Age': age})


print("Age Mean and Median Before Outliers")
print(" Mean: ", df['Age'].mean(),"\n","Median: ", df['Age'].median()) 

df=pd.concat([df, pd.DataFrame({'Age': [200, 150, -170, 300]})], ignore_index=True)

#Histogram
plt.hist(df['Age'],bins=20, edgecolor='Black')
plt.show()

#Box Plot
plt.boxplot(df['Age'])
plt.show()

print("Age Mean and Median After Outliers")
print(" Mean: ", df['Age'].mean(),"\n","Median: ", df['Age'].median(),"\n")
print("- Outliers significantly increased the mean and median remain the same")
```

Q.2 : Outlier Detection and Removal using the Z-score Method

Theory Recap:

- Z-score measures how many standard deviations a value is away from the mean. 2 = x-u/o
- Typically, IZ|>3 → Outlier.

Tasks:
- Use the dataset with outliers from Task 1.
- Compute mean (u) and standard deviation (o).
- Calculate Z-score for each data point.
- Identify and remove points with |Z| > 3.
- Plot cleaned dataset vs original.

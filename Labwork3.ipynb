Que 1:- EDA
Theory Recap:
EDA is the process of understanding, summarizing, and visualizing datasets before
modeling.
- Identify distributions
- Detect anomalies/outliers
Check correlations
Summarize key statistics
Tasks:
- Load a dataset (Iris / Titanic / any CSV).
- Display shape, data types, missing values, and descriptive stats.
Create a summary table with mean, median, mode for numeric features.
Plot histograms for all numeric features.

```python
import pandas as pd
import matplotlib.pyplot as plt

#load dataset
df=pd.read_csv("students.csv")

#first 5 rows
print(df.head(),"\n")

#Shape of data set
print("Shape of Dataset:", df.shape,"\n")

#Data Types
print("Data Types of Dataset:","\n" ,df.dtypes,"\n")

#Missing Values 
print("Missing Values: \n", df.isna().sum(), "\n")

#Summary stats
print("Summary: \n", df.describe(),"\n")

#Create a summary table for numeric colums

numeric_df = df.select_dtypes(include="number")

summary = pd.DataFrame({
    "Mean": numeric_df.mean(),
    "Median": numeric_df.median(),
    "Mode": numeric_df.mode().iloc[0]
})

print(summary)

#Histogram Graph for numeric columns
numeric_df.hist()
plt.show()

```

Que 2:- EDA using Univariate Analysis
Theory Recap:
Univariate = analyzing one variable at a time.
For cutegorical i treguency our, bar chary tats
For categorical → frequency count, bar chart
Tasks:
- Choose one numeric column (e. g, Age). Plot histogram and boxplot.
Calculate skewness and kurtosis.
Choose one categorical column (e.g., Gender). Plot bar chart of counts.

```python
# 1. Numeric Column (Marks)

# Histogram
plt.hist(df["Marks"])
plt.title("Histogram of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()

# Boxplot
plt.boxplot(df["Marks"])
plt.title("Boxplot of Marks")
plt.show()

# Skewness & Kurtosis
print("Skewness:", df["Marks"].skew())
print("Kurtosis:", df["Marks"].kurt())

# 2. Categorical Column (Create Grade)

# Create Grade column
df["Grade"] = pd.cut(df["Marks"], bins=[0, 60, 80, 100], labels=["C", "B", "A"])

# Count values
print("\nGrade Counts:\n", df["Grade"].value_counts())

# Bar chart
df["Grade"].value_counts().plot(kind="bar")
plt.title("Grade Distribution")
plt.xlabel("Grade")
plt.ylabel("Count")
plt.show()
```

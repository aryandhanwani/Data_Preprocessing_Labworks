Que 1:- Missing Indicator + Random Sample Imputation

Theory Recap:
Sometimes missingness itself carries information.
- We create a missing indicator column (0 = value present, 1 = missing).
- Then fill missing values with a random sample from the same column.

Tasks:
- Create a dataset with missing values in Age and Salary.
- Add a new column Age_missing that indicates whether Age was missing.
- For missing Age values, randomly sample from available values.
- Compare original vs imputed dataset.

```python
import pandas as pd
import numpy as np

data = {
    "Age": [20, 25, np.nan, 30, np.nan, 35],
    "Salary": [30000, 40000, 50000, np.nan, 45000, 60000]
}

df = pd.DataFrame(data)

#Creating new columns in numeric 
df["Age_Missing"] = np.where(df["Age"].isna(), 1, 0)

print(df,"\n")

print("Comparission with Imputing Values: \n")

# Count missing values
n_missing = df["Age"].isna().sum()

# Get random samples from existing values
random_values = df["Age"].dropna().sample(n_missing, random_state=1)

# Fill missing values
df.loc[df["Age"].isna(), "Age"] = random_values.values

print(df)

```

Que 2:- KNN Imputer (Multivariate Imputation)

Theory Recap:
KNNImputer uses similar rows (neighbors) to predict missing values.
: Works well when features are correlated.
Example: If Age is missing, we can impute based on Salary and Experience of
nearest neighbors.

Tasks:
- Create a dataset with missing values in multiple columns.
- Use KNNImputer from sklearn impute with n_ neighbors=3.
- Compare dataset before and after imputation.
- Experiment with different values of k and compare results.

```python
from sklearn.impute import KNNImputer

print("Original Dataset:\n")
print(df,"\n")

imputing=KNNImputer(n_neighbors=3)

df_filled2=pd.DataFrame(imputing.fit_transform(df[['Salary']]),columns=['Salary'])

print("After Imputing Data: \n")
print(df_filled2)
```

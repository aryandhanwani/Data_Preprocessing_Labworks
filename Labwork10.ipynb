Q.1 : Data Ordinal Encoding

Theory Recap:
Used when categorical variables have a natural order (Low < Medium < High)
- Assigns integer values based on order.

Tasks:
- Create a dataset with column Education containing: [High School, Graduate,
Postgraduate].
- Define an order mapping (High School-1, Graduate-2, Postgraduate=3).
Encode values into numeric form using mapping or OrdinalEncoder.
- Verify encoding by checking order is preserved.

```python
import pandas as pd

# Step 1: Create dataset
df = pd.DataFrame({
    'Education': ['High School', 'Graduate', 'Postgraduate', 'Graduate', 'High School']
})

print("Original:\n", df)

mapping = {
    'High School' : 0,
    'Graduate' : 1,
    'Postgraduate' : 2 }

#Apply encoding
df['Educaition_mapping'] = df['Education'].map(mapping)

print("\nAfter Mapping:\n", df)

```

Q.2 : Label Encoding

Theory Recap:
Assigns an integer to each category.
No notion of order (just IDs).
- Works for target variable or categorical features in tree-based models.

Tasks:
- Create a dataset with column City containing: [Delhi, Mumbai, Chennai,
Kolkata].
- Apply LabelEncoder from sklearn.preprocessing.
- Print encoded values.
- Try decoding back from numbers to original labels.

```python
from sklearn.preprocessing import LabelEncoder

# Step 1: Create dataset
df = pd.DataFrame({
    'City': ['Delhi', 'Mumbai', 'Chennai', 'Kolkata', 'Mumbai', 'Delhi']
})

print("Original:\n", df)

# Step 2: Create ONE encoder object
le = LabelEncoder()

# Step 3: Encode (use 1D column)
df['City_label'] = le.fit_transform(df[['City']])

print("\nAfter Label Encoding:\n", df)

# Step 4: Decode using SAME encoder
decoded = le.inverse_transform(df['City_label'])

print("\nDecoded:\n", decoded)
```

Q.3 : One-Hot Encoding

Theory Recap:
- Creates separate binary columns for each category.
- Prevents false ordering issues of Label Encoding
Increases dimensionality (sparse representation).

Tasks:
Create a dataset with column Department containing: [HR, IT, Finance, Sales].
- Apply One-Hot Encoding using:
- pd.get_dummies()
- OneHotEncoder from skleam.preprocessing.
- Compare output from both methods.
- Discuss which encoding is better for linear models vs tree-based models.

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
df = pd.DataFrame({
    'Department': ['HR', 'IT', 'Finance', 'Sales', 'IT', 'HR']
})
print("Original:\n", df)
#Step 1 : get dummies
df_dummies = pd.get_dummies(df['Department'])
print("\nAfter get_dummies:\n", df_dummies)

# Step 2: OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)

encoded = encoder.fit_transform(df[['Department']])

df_encoded = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())

print("\nAfter OneHotEncoder:\n", df_encoded)
```

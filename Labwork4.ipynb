Que 1:- Theory Recap:
- Real-world datasets often contain missing values (NaN).
- Imputation = replacing missing values with meaningful substitutes.
- Strategies: Mean, Median, Mode, Most Frequent, Constant.

Tasks:
- Load a dataset (Titanic / custom CSV) with missing values.
- Check missing values using .isnullO.sum0.
- Drop rows/columns with too many missing values.
- Apply imputation techniques to fill missing values.

```python
import pandas as pd

data = {
    "Name": ["Aryan", "Rohit", "Neha", "Priya", "Aman"],
    "Age": [20, None, 19, 21, None],
    "Marks": [75, 60, None, 90, 65]
}

df = pd.DataFrame(data)
print(df.head(),"\n")

#checking missing values
print(df.isnull().sum(),"\n")

#dropping null value
df = df.dropna()

#Filling null values
df["Age"].fillna(df["Age"].mean())
df["Marks"].fillna(df["Marks"].median())
```

Que 2:- Theory Recap:
- sklearn impute SimpleImputer can fill missing numeric values with mean,
median, or a constant.

Tasks:
Create a DataFrame with missing numeric values in columns.
Use SimpleImputer(strategy="mean") to replace with column mean.
Repeat using strategy-"median".
Compare before vs after imputation.

```python
from sklearn.impute import SimpleImputer

#Data loading
data = {
    "Age": [20, 25, None, 30, None],
    "Marks": [80, None, 70, 60, 90]
}

df = pd.DataFrame(data)
print("Original Data:\n", df)

mean_imputer = SimpleImputer(strategy="mean")

df_mean = pd.DataFrame(
    mean_imputer.fit_transform(df),
    columns=df.columns
)

print("\nAfter Mean Imputation:\n", df_mean)

median_imputer = SimpleImputer(strategy="median")

df_median = pd.DataFrame(
    median_imputer.fit_transform(df),
    columns=df.columns
)

print("\nAfter Median Imputation:\n", df_median)
```

Que 3:- Theory Recap:
Categorical values can be imputed using:
- Mode (most frequent category)
- A placeholder like "Unknown"

Tasks:
- Create a DataFrame with missing values in categorical column (e.g., Gender,
Department).
- Use SimpleImputer(strategy="most _frequent") to fill missing values with the
most common category.
- Use Simplemputer(strategy="constant", fill_ value="Unknown") to replace
with placeholder.

```python
data = {
    "Name": ["Aryan", "Rohit", None, "Aryan", "Aman"],
    "City": ["Surat", None, "Surat", "Delhi", None]
}

df = pd.DataFrame(data)
print("Original Data:\n", df)

imputer = SimpleImputer(missing_values=None,strategy="most_frequent")

df_filled = pd.DataFrame(
    imputer.fit_transform(df),
    columns=df.columns
)

print("\nAfter Imputation:\n", df_filled)
```

Que 4:- Theory Recap
Useful when missing values should be replaced with the most common class.
- Common in categorical data like Gender, City, Occupation.

Tasks:
- Create a dataset with missing categorical values (e.g., Cities: Delhi, Mumbai,
NaN, NaN, Delhi).
- Apply SimpleImputerstrategy-"most _frequent".
- Print the dataset before & after imputation.
- Compare results when using "constant" vs "most _frequent".

```python
from sklearn.impute import KNNImputer

data = {
    "Age": [20, 25, None, 30, 35],
    "Marks": [80, None, 70, 60, 90]
}

df = pd.DataFrame(data)
print("Original Data:\n", df)

imputer = KNNImputer(n_neighbors=2)

df_filled = pd.DataFrame(
    imputer.fit_transform(df),
    columns=df.columns
)

print("\nAfter KNN Imputation:\n", df_filled)

```
